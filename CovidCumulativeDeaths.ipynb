```python
# This downloads some Covid data
import urllib
result = urllib.request.urlretrieve('https://covid.ourworldindata.org/data/owid-covid-data.csv', 'owid-covid-data.csv')
```

```python
# This reads it into a pandas data frame
import pandas as pd
covid = pd.read_csv('owid-covid-data.csv', parse_dates=['date'])
# This shows the columns in the data
covid.columns
```

```python
import matplotlib.pyplot as plt

OneCountry = covid.loc[covid['iso_code'] == 'GBR'] # This selects one country

plt.figure(figsize=(10,6))  # make it a nice size

plt.plot(OneCountry['date'], OneCountry['total_deaths_per_million'])
plt.show()
```

```python
# This is a more complicated example that lets you compare different countries
countries_to_plot = ['GBR', 'USA', 'FRA', 'ITA', 'PER', 'ISR']  # This is just a list of countries

# The code here filters the covid data using a method called query
# The result is grouped by iso_code (which is the country)
countries = covid.query('iso_code in ' + str(countries_to_plot)).groupby('iso_code')
countries = covid.groupby('iso_code')
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
for name, country in countries:
  plt.plot(country['date'], country['total_deaths_per_million'], label=name,  marker="o", linestyle="")
plt.xticks(rotation=90)
plt.xlabel('Date')
plt.ylabel('Deaths per million')
plt.title('Cumulative Covid 19 deaths per million')
#plt.legend()
plt.show()
```
